### `df.info()` in Pandas

- `df.info()` shows a **quick summary** of the DataFrame.
- It includes:
  - Number of **rows** and **columns**
  - **Column names**
  - **Data types** of each column
  - Number of **non-null (non-missing)** values per column

```python
print(df.columns)
print("Total Number of Columns: ", len(df.columns))
print("Total Number of Rows: ", len(df))
print(df.shape)
```

### `df.describe()` in Pandas (Integer Data Only)

- `df.describe()` generates **descriptive statistics** for **numeric (integer)** columns.
- It includes:
  - **count** → Total number of non-null values
  - **mean** → Average value
  - **std** → Standard deviation (spread of the data)
  - **min** → Minimum value
  - **25%** → 1st quartile (25th percentile)
  - **50%** → 2nd quartile (Median, 50th percentile)
  - **75%** → 3rd quartile (75th percentile)
  - **max** → Maximum value

```python
df_Country=pd.read_excel('/content/drive/MyDrive/Project and Coding/Exploratory Data Analysis (EDA)/Country-Code.xlsx')
```

***Merging same columns from two dataframes***

---


```python
from typing_extensions import final
final_df=pd.merge(df,df_Country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

**It will return a Series where the index is the unique values in the Country column (i.e., country names) and the values are the count of how many times each country appears in that column.**




```python
final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
country_val=final_df.Country.value_counts()
```

```python
# ploting pie chart for top three country's that uses zomato
plt.pie(x=country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

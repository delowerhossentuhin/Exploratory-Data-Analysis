### `df.info()` in Pandas

- `df.info()` shows a **quick summary** of the DataFrame.
- It includes:
  - Number of **rows** and **columns**
  - **Column names**
  - **Data types** of each column
  - Number of **non-null (non-missing)** values per column

```python
df.info()
```

```python
print(df.columns)
print("Total Number of Columns: ", len(df.columns))
print("Total Number of Rows: ", len(df))
print(df.shape)
```

### `df.describe()` in Pandas (Integer Data Only)

- `df.describe()` generates **descriptive statistics** for **numeric (integer)** columns.
- It includes:
  - **count** → Total number of non-null values
  - **mean** → Average value
  - **std** → Standard deviation (spread of the data)
  - **min** → Minimum value
  - **25%** → 1st quartile (25th percentile)
  - **50%** → 2nd quartile (Median, 50th percentile)
  - **75%** → 3rd quartile (75th percentile)
  - **max** → Maximum value

```python
df_Country=pd.read_excel('/content/drive/MyDrive/Project and Coding/Exploratory Data Analysis (EDA)/Country-Code.xlsx')
```

```python
df_Country.head()
df_Country.shape
```

***Merging same columns from two dataframes***

---


```python
final_df=pd.merge(df,df_Country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

**It will return a Series where the index is the unique values in the Country column (i.e., country names) and the values are the count of how many times each country appears in that column.**




```python
final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
country_val=final_df.Country.value_counts()
```

```python
# ploting pie chart for top three country's that uses zomato
plt.pie(x=country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

##Observations

1.   When Rating Between 4.5 to 4.9--->Excellent
2.   When Rating Between 4.0 to 4.4--->Very Good
3.   When Rating Between 3.5 to 3.9--->Good
4.   When Rating Between 2.5 to 3.4--->Average
5.   When Rating Between 0.0 to 2.4-->Poor
---
##Let's Draw this Observation

```python
ratings.head()
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

##Observation

#**Ques-1: Find the countries that has given 0 rating?**


---



```python
final_df[final_df['Rating color']=='White'].groupby(['Aggregate rating','Country']).size().reset_index().rename(columns={0:"Total Count"})
```

```python
ratingWithZero=final_df[final_df['Rating color']=='White'].groupby(['Aggregate rating','Country']).size().reset_index().rename(columns={0:"Total Count"})
sns.barplot(x='Country',y='Total Count',data=ratingWithZero)
```

# **Ques-5: Find the top ten Cusinies?**

```python
topTenCuisines=final_df.groupby('Cuisines').size().reset_index().rename(columns={0:"Total"}).sort_values(by='Total',ascending=False).head(10)
```

```python
sns.barplot(x='Cuisines',y='Total',data=topTenCuisines)
```
